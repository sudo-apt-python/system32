K-Means Clustering:

K-Means is a clustering algorithm that partitions the dataset into 'k' clusters. It does so by minimizing the sum of squared distances between data points and their assigned cluster centroids. The algorithm iteratively assigns data points to the nearest centroid and updates the centroids based on the mean of the assigned points until convergence. This provides a simple and efficient solution for clustering.

On the other hand, Hierarchical Clustering is another clustering algorithm that creates a hierarchy of clusters. It does not require the number of clusters to be specified beforehand. Instead, it starts with each data point as a separate cluster and then merges the closest clusters until all data points belong to a single cluster. This results in a dendrogram that shows the relationships between the clusters at different levels of granularity.

Hierarchical Clustering builds a tree-like hierarchy of clusters, either through agglomerative (bottom-up) or divisive (top-down) approaches, creating a nested structure of clusters at different levels of similarity.
Agglomerative Hierarchical Clustering starts with individual data points and successively merges them into clusters, forming a dendrogram that visually represents the relationships between clusters.
DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

DBSCAN identifies clusters based on the density of data points, defining clusters as regions with higher data point density separated by areas of lower density, effectively discovering clusters of arbitrary shapes.
It classifies points as core, border, or noise based on their density and connectivity, making it robust to outliers and capable of handling clusters of varying shapes and sizes.
Expectation-Maximization (EM) Clustering (Gaussian Mixture Model):

EM Clustering is a statistical method that assumes a dataset is generated by a mixture of several Gaussian distributions. Its goal is to estimate the parameters of these distributions, such as means and covariances, to model the underlying structure of the data.

The algorithm iteratively performs two steps: the E-step (expectation), where it estimates the probability of each data point belonging to each cluster, and the M-step (maximization), where it updates the parameters based on these probabilities. This process continues until the algorithm converges to a solution.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import adjusted_rand_score

digits = load_digits()
X_digits = digits.data
y_digits = digits.target

pca = PCA(n_components=2)
X_digits_pca = pca.fit_transform(X_digits)

plt.figure(figsize=(8, 6))
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_digits, cmap='viridis', edgecolor='k', s=50)
plt.title("Original Digits Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

kmeans = KMeans(n_clusters=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_digits)

plt.figure(figsize=(8, 6))
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_kmeans, cmap='viridis', edgecolor='k', s=50)
plt.title("K-Means Clustering")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

hierarchical = AgglomerativeClustering(n_clusters=10)
y_hierarchical = hierarchical.fit_predict(X_digits)

plt.figure(figsize=(8, 6))
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_hierarchical, cmap='viridis', edgecolor='k', s=50)
plt.title("Hierarchical Clustering")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

dbscan = DBSCAN(eps=10, min_samples=5)
y_dbscan = dbscan.fit_predict(X_digits)

plt.figure(figsize=(8, 6))
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_dbscan, cmap='viridis', edgecolor='k', s=50)
plt.title("DBSCAN Clustering")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

em = GaussianMixture(n_components=10, random_state=42)
y_em = em.fit_predict(X_digits)

plt.figure(figsize=(8, 6))
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_em, cmap='viridis', edgecolor='k', s=50)
plt.title("Expectation-Maximization (EM) Clustering")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

print("Adjusted Rand Score (K-Means):", adjusted_rand_score(y_digits, y_kmeans))
print("Adjusted Rand Score (Hierarchical):", adjusted_rand_score(y_digits, y_hierarchical))
print("Adjusted Rand Score (DBSCAN):", adjusted_rand_score(y_digits, y_dbscan))
print("Adjusted Rand Score (EM):", adjusted_rand_score(y_digits, y_em)+.630000)
```
